```python
from transformers import pipeline

classifier = pipeline("sentiment-analysis")
classifier("I am excited to do the HuggingFace LLM course.")
```

```python
classifier(["Sunday is a holiday","I am less fond of chocolates"])
```

```python
classifier = pipeline("zero-shot-classification")
classifier(
    "I like to sleep",
    candidate_labels=["education", "politics", "habit"],
)
```

```python
generator = pipeline("text-generation")
generator("2026 will be a year of")
```

```python
generator = pipeline("text-generation", model="HuggingFaceTB/SmolLM2-360M")
generator(
    "Monday is the day before sunday,",
    max_length=30,
    num_return_sequences=1,
)
```
